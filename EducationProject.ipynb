## Question 1: Does a state’s average fourth-grade test scores predict its average eighth-grade scores in reading and math?

### Modeling Approach: 
We will approach this question by analyzing the linear regression models of the relationships between 4th and 8th grade math scores, 4th and 8th grade reading scores, 4th grade math and reading scores, and 8th grade math and reading scores.

```python
# drop null values from all pertinent columns in a new dataframe
df_q1 = df[['STATE', 'YEAR', 'AVG_MATH_4_SCORE', 'AVG_MATH_8_SCORE', 
            'AVG_READING_4_SCORE', 'AVG_READING_8_SCORE']].copy()
df_q1 = df_q1[df_q1.STATE.notnull()]
df_q1 = df_q1[df_q1.YEAR.notnull()]
df_q1 = df_q1[df_q1.AVG_MATH_4_SCORE.notnull()]
df_q1 = df_q1[df_q1.AVG_MATH_8_SCORE.notnull()]
df_q1 = df_q1[df_q1.AVG_READING_4_SCORE.notnull()]
df_q1 = df_q1[df_q1.AVG_READING_8_SCORE.notnull()]
df_q1.head(10)
```

### Linear Regression Model - 4th Grade Math vs. 8th Grade Math

```python
# import useful packages
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# split the data into training and test sets
X, y = df_q1.AVG_MATH_4_SCORE, df_q1.AVG_MATH_8_SCORE
X_train, X_test, y_train, y_test = train_test_split(np.reshape(X, (-1, 1)), y, 
                                                    test_size=0.3, random_state=4)
```

```python
# build a linear regressor
lin = LinearRegression().fit(X_train, y_train)
lin_y_pred = lin.predict(X_test)

# get statistics for the model
print("Linear Model Accuracy (r^2): ", lin.score(X_test, y_test))
print()
print("MSE: ", mean_squared_error(y_test, lin_y_pred))
print()
print("Coefficients: ", lin.coef_)
print()
print("Intercept: ", lin.intercept_)
```

```python
import matplotlib.pyplot as plt

# plot actual vs. preds
fig,ax = plt.subplots()
ax.scatter(y_test, lin_y_pred)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
plt.title('Actual vs. Predicted (4th Grade Math vs. 8th Grade Math)')
fig.show()
```

## Question 2: What is the relationship between state finances (revenue and expenditure) and the strength of student test scores?

### Modeling Approach: 
For this question, we plan cluster total revenue and total expenditure against an average score metric across the two grades and disciplines. Then, we will evaluate the cluster models and their significance to the question.

```python
# filter dataframe
from sklearn.preprocessing import StandardScaler 

data_filtered = df[['TOTAL_REVENUE', 'TOTAL_EXPENDITURE', 'AVG_MATH_4_SCORE', 
                               'AVG_MATH_8_SCORE', 'AVG_READING_4_SCORE', 'AVG_READING_8_SCORE']]

data_filtered = data_filtered[data_filtered.TOTAL_REVENUE.notnull()]
data_filtered = data_filtered[data_filtered.TOTAL_EXPENDITURE.notnull()]
data_filtered = data_filtered[data_filtered.AVG_MATH_4_SCORE.notnull()]
data_filtered = data_filtered[data_filtered.AVG_MATH_8_SCORE.notnull()]
data_filtered = data_filtered[data_filtered.AVG_READING_4_SCORE.notnull()]
data_filtered = data_filtered[data_filtered.AVG_READING_8_SCORE.notnull()]

total_score = data_filtered[['AVG_MATH_4_SCORE', 'AVG_MATH_8_SCORE', 'AVG_READING_4_SCORE', 
                             'AVG_READING_8_SCORE']]

data_filtered['AVERAGE_SCORE'] = total_score.sum(axis = 1) / 4
s = StandardScaler()
data_scaled = s.fit_transform(data_filtered)
data_scaled.shape
data_filtered.head(10)
data_filtered.apply(np.log)
```

```python
# initialize the first cluster model with KMeans
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from matplotlib import pyplot as plt

cluster_model1 =  KMeans(n_clusters=4) 
cluster_model1.fit(X=data_scaled)
clusters1 = cluster_model1.predict(data_scaled)
```

## Question 3: Is student diversity in race predictive of average test scores by state?

### Modeling Approach: 
We plan to calculate a metric (Simpson's Diversity Index) to measure diversity for each state-year given race statistics included in the dataset. Then, we will look at the strength of relationships between diversity score and math and reading test scores for 4th and 8th grade through a heatmap. If any seem significant, we will evaluate that significance with Stats Model API.


```python
# drop null values from all pertinent columns in a new dataframe
# Key: AM (American Indian or Alaska Native), AS (Asian), HI (Hispanic/Latino), 
# BL (Black or African American), WH (White), HP (Hawaiian Native/Pacific Islander), and 
# TR (Two or More Races).

df_q3 = df[['STATE', 'YEAR', 'GRADES_ALL_AM', 'GRADES_ALL_AS', 'GRADES_ALL_HI', 'GRADES_ALL_BL',
            'GRADES_ALL_WH', 'GRADES_ALL_HP', 'GRADES_ALL_TR', 'GRADES_ALL_G', 
            'AVG_MATH_4_SCORE', 'AVG_MATH_8_SCORE', 'AVG_READING_4_SCORE', 
            'AVG_READING_8_SCORE']].copy()
df_q3 = df_q3[df_q3.GRADES_ALL_AM.notnull()]
df_q3 = df_q3[df_q3.GRADES_ALL_AS.notnull()]
df_q3 = df_q3[df_q3.GRADES_ALL_HI.notnull()]
df_q3 = df_q3[df_q3.GRADES_ALL_BL.notnull()]
df_q3 = df_q3[df_q3.GRADES_ALL_WH.notnull()]
df_q3 = df_q3[df_q3.GRADES_ALL_HP.notnull()]
df_q3 = df_q3[df_q3.GRADES_ALL_TR.notnull()]
df_q3 = df_q3[df_q3.GRADES_ALL_G.notnull()] 
df_q3 = df_q3[df_q3.AVG_MATH_4_SCORE.notnull()]
df_q3 = df_q3[df_q3.AVG_MATH_8_SCORE.notnull()]
df_q3 = df_q3[df_q3.AVG_READING_4_SCORE.notnull()]
df_q3 = df_q3[df_q3.AVG_READING_8_SCORE.notnull()]
df_q3.head(10)
```

#### Calculate a diversity metric through Simpson's Diversity Index
Below information about the index via: https://www.statisticshowto.datasciencecentral.com/simpsons-diversity-index/

"Simpson’s diversity index (SDI) measures community diversity. Although it’s commonly used to measure biodiversity, it can also be used to gauge diversity differences in populations in schools, communities and other locations.

The range is from 0 to 1, where:

+ High scores (close to 1) indicate high diversity.
+ Low scores (close to 0) indicate low diversity.

One of more the useful aspects of the index is to compare two sets of data to see which is more diverse. For example, if one has an SDI of 0.5 and another has an SDI of 0.35, then the set with the SDI of 0.5 is more diverse."

```python
# calculate Simpson's Diversity Index and add it to the dataframe
def numerator(lst):
    sum = 0
    for column in lst:
        sum += column * (column - 1)
    return sum
    

numerator = numerator([df_q3['GRADES_ALL_AM'], df_q3['GRADES_ALL_AS'], df_q3['GRADES_ALL_HI'], 
                               df_q3['GRADES_ALL_BL'], df_q3['GRADES_ALL_WH'], df_q3['GRADES_ALL_HP'], 
                               df_q3['GRADES_ALL_TR']])
denominator = df_q3['GRADES_ALL_G'] * (df_q3['GRADES_ALL_G'] - 1)
df_q3['DIVERSITY_SCORE'] = 1 - (numerator / denominator)                      
```

## Executive Summary

#### Question 1:
Our first question asks whether 4th grade scores in math and reading are predictive of 8th grade scores in the subjects. Our modeling approach to this question yielded interesting results. We found a stark difference between the R^2 values of linear regression models for 4th vs. 8th grade math scores and 4th grade math vs. 4th grade reading scores (very high positive correlation) and 4th vs. 8th grade reading scores and 8th grade math vs. 8th grade reading scores (very low positive / no correlation). Our first model **found that 4th grade math scores explain about 85% of the variability in 8th grade math scores wihtin our dataset.** The other significant model **found that 4th grade math scores explain about 73% of the variability in 4th grade reading scores within our dataset**. 

Relationships between 4th and 8th grade reading and 8th grade reading and math were not found significant given low R^2 scores of 0.007 and 0.015, respectively. 

Plausible explanations for these findings could include the nature of mathematics as subject which is more trainable; states with better systems in place may see high scores across all grade levels. Reading varies more with the individual child's reading habits and natural reading ability. Additionally, 4th grade scores are more highly correlated than 8th grade scores across topics, possibly because in 4th grade there is not yet a divide of CP, honors, and accelerated classes which may accelerate the levels of certain students and not others.

#### Question 2:

To answer question two, we created two cluster plots: one plotting average score for each state-year against total expenditure and the other plotting each state-year against total revenue. We formed four clusters in each model. Interestingly, the two cluster models were nearly identical. This suggests that revenue and expenditure are highly correlated, a relationship we verified in the linear regressor below the scatter plots with a 99% R^2 value, and there is little-to-no difference between these the relationships of these two variables and average test scores across the two grades and disciplines.

The cluster tiers appear to be:
+ Low expenditure/revenue, up to 240 average score
+ Low expenditure/revenue, 240-250 average score
+ Low expenditure/revenue, 250+ average score
+ High expenditure/revenue, medium average score

Ineterestingly, the fewer number of states that earn and spend significantly higher amounts on their school systems do not see a matching payoff in scores according to this dataset. The two largest clusters are low expenditure/revenue and 240-250 and 250+ average scores, respectively.

#### Question 3:

Our third question examines student diversity in race (Simpon's Diversity Index) against average test scores in 4th and 8th grade and math and reading. After creating a heatmap, we saw potentially significant relationships between diversity score and, particulary, 4th grade math and reading scores. Completing a Stats Model API analysis of the relationship between these two variables and diversity score, we found that the p-score and R^2 values for these relationships were not significant. Thus, our findings do not suggest that diversity influences test scores directly.
